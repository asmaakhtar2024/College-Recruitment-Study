### Question 1 Compare the age demographics of the survey over time (2020-2024).  What are the changes over time, if any?

```python
#read in files for different years
survey20="fall2020non-majors.csv"
#survey21="fall2021non-majors.csv"
survey22="fall2022non-majors.csv"
survey23="fall2023non-majors.csv"
survey24="fall2024non-majors.csv"
# save files to dataframe 
df2020=pd.read_csv(survey20,index_col=0)
#df2021=pd.read_csv(survey21,index_col=0)
df2022=pd.read_csv(survey22,index_col=0)
df2023=pd.read_csv(survey23,index_col=0)
df2024=pd.read_csv(survey24,index_col=0)
```

```python
#Save age counts into new data set for each year
df2020 = df2020['Age '].value_counts().reset_index() 
df2022 = df2022['Age '].value_counts().reset_index() 
df2023 = df2023['Age '].value_counts().reset_index()
df2024 = df2024['Age '].value_counts().reset_index() 
#Rename age columns for each year
df2020 = df2020.rename(columns={'Age ': 'age'})
df2022 = df2022.rename(columns={'Age ': 'age'})
df2023 = df2023.rename(columns={'Age ': 'age'})
df2024 = df2024.rename(columns={'Age ': 'age'})
#Create new columns for age counts for each year
df2020.columns = ['age', 'count'] 
df2022.columns = ['age', 'count'] 
df2023.columns = ['age', 'count'] 
df2024.columns = ['age', 'count'] 
#Create new columns for different years
df2020['year'] = 2020
df2022['year'] = 2022
df2023['year'] = 2023
df2024['year'] = 2024
```

```python
#Display 2020 data
df2020
```

```python
#Display 2022 data
df2022
```

```python
#Display 2023 data
df2023
```

```python
#new data frame for 2024
df2024
```

```python
#concatenate data frames together
df_ages=pd.concat([df2020,df2022,df2023,df2024])
#diplay the combined data frames 
df_ages
```

### Grouped barchart for age demographics

```python
#source chatgpt to make the graph
#pivot data so that 'age' is the index and each year is a column 
df_pivot = df_ages.pivot_table(index='age', columns='year', values='count', fill_value=0) 
#create grouped bar chart
df_pivot.plot(kind='bar', figsize=(10, 6), width=0.8) 
#titles and labels 
plt.title('Age Demographics From 2020-2024') 
plt.xlabel('Age') 
plt.ylabel('Number of Students') 
#rotate x-axis labels for clarity
plt.xticks(rotation=45, ha='right') 
#place the legend outside the plot area
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
#display the graph 
plt.show()
```

## Question #2 What are the top 5 places students heard about CCM?

```python
#Display column names 
df.columns
```

```python
#Filter out columns beginning with heard
heard_columns=df.filter(like='heard').columns
heard_columns
```

```python
#add up amount of 'yes' responses for each column
heard_yes = df[heard_columns].apply(lambda x: (x=='Yes').sum())
heard_yes
```

```python
#Sort in the descending order
heard_sorted = heard_yes.sort_values(ascending=False)
heard_sorted
```

```python
#plot the data
plt.figure(figsize=(10, 6))
heard_sorted.plot(kind='barh') 
#labels and title 
plt.title('Most and Least Common Ways People Heard About College')
plt.xlabel('Yes Responses')
plt.ylabel('Ways Students Heard About CCM') 
#display the plot
plt.show()
```

```python
#Display top 5 heard responses
heard_top5=heard_sorted.head(5)
heard_top5
```

## Question #4 Rank all the student responses that impacted their decision to attend County College of Morris from the most popular to least popular.

```python
#Filter out columns beginning with impact
impact_columns=df.filter(like='impact').columns
#Display data
impact_columns
```

```python
# Turn the data frame into series to get value counts
impact_counts = df[impact_columns].apply(pd.Series.value_counts).transpose() #transpose to make column header amount of impact
impact_counts
```

```python
#Reindex the columns from highest to lowest impact 
impact_counts = impact_counts.reindex(['High Impact', 'Some Impact', 'No Impact', 'N/A'], axis=1)
impact_counts
```

```python
#Create a stacked bar chart
impact_counts.plot(kind='barh', stacked=True)

#title and label the chart
plt.title('Extent the Following things Impacted Students Decision to Attend CCM')
plt.xlabel('Amount of Responses')
plt.ylabel('Impacted Factors')


#place the legend outside the plot area
plt.legend(title='Amount of Impact', bbox_to_anchor=(1.05, 1), loc='upper left')

#display the graph
plt.show()
```
